```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Load data to pandas to gather info in dataframe

```python
ccd = pd.read_csv("/content/creditcard.csv")
```

```python
ccd.head()
```

```python
ccd.tail()
```

```python
ccd.info()
```

```python
ccd.isnull().sum()
```

```python
ccd['Class'].value_counts()
```

```python
legit = ccd[ccd.Class == 0]
fraud = ccd[ccd.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Statistical measrures of the data


```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
ccd.groupby('Class').mean()
```

Undersampling of data


```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets

```python
x = new_dataset.drop(columns = 'Class', axis=1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Spliting the data into training data and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state=2)
```

```python
print (x.shape, x_train.shape, x_test.shape)
```

```python
print (y.shape, y_train.shape, y_test.shape)
```

Model training using logistic regression

```python
model = LogisticRegression()
```

Training the logistic regression model with the training data

```python
model.fit(x_train, y_train)
```

Model evaluation : Accuracy score

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy on training data : ", training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy on test data : ", test_data_accuracy)
```
